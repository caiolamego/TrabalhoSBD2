```python
import sdmx
import pandas as pd


# Parâmetros de tempo e atributos
params = dict(
    startPeriod="2000"
)

# Cliente IMF
client = sdmx.Client("IMF_DATA")

base = client.dataflow('PCPS')

estrutura = base.structure['DSD_PCPS']
indicadores = base.codelist['CL_PCPS_INDICATOR']
indicadores_commodities = indicadores.items

df_final = pd.DataFrame() 

for i,indicador in enumerate(indicadores_commodities):

    key = f'BRA+USA+CHN+IND.{indicador}.USD.A'

    # Baixar os dados
    data_msg = client.data("PCPS", key=key, params=params, dsd=estrutura)

    # Converter para pandas DataFrame
    df = sdmx.to_pandas(
        data_msg, 
        #datetime="TIME_PERIOD"
        )

    df = df.reset_index()

    df_final = pd.concat([df_final, df], ignore_index=True)
    print(f"Finalizada etapa {i+1}")




display(df_final.head())

df_final.to_csv("bases_dados/commodities.csv", index=False)
```

```python
import sdmx

client = sdmx.Client("IMF_DATA")

# 1) Ver fluxos e DSD
flow = client.dataflow("PCPS")
dsd_id = list(flow.structure.keys())[0]  # deve incluir 'DSD_PCPS'
dsd = client.datastructure(dsd_id).datastructures[dsd_id]

# 2) Ver dimensões e codelists
dims = [c.id for c in dsd.dimensions.components]
print("Dimensões:", dims)
for comp in dsd.dimensions.components:
    if comp.local_representation.enumerated is not None:
        cl = comp.local_representation.enumerated
        print(comp.id, "→ codelist:", cl.id)
        for item in list(cl.items())[:20]:  # amostra
            print(" ", item[0], "-", item[1].name)

# 3) Monte a chave respeitando a ORDEM acima e usando códigos válidos
# Exemplo ilustrativo: key = '<DIM1>.<DIM2>.<DIM3>...'
# NUNCA inclua 'USA' (PCPS não tem país). Use COMMODITY / PRICE_TYPE / CURRENCY / FREQ conforme a DSD.
```

```python
import sdmx

# 1) Clientes separados
imf_struct = sdmx.Client("IMF")        # estruturas (sem dados)
imf_data   = sdmx.Client("IMF_DATA")   # dados

# 2) Descobrir o DSD do fluxo PCPS
sm_flow = imf_struct.dataflow("PCPS")
df_def = sm_flow.dataflows["PCPS"]
dsd_id = df_def.structure.id  # ex.: 'DSD_PCPS'
print("DSD id:", dsd_id)

# 3) Baixar a DSD com descendentes (codelists etc.)
sm_dsd = imf_struct.datastructure(dsd_id, params=dict(references="descendants"))

# Acesso correto: 'data_structures' (não 'datastructures')
dsd = sm_dsd.data_structures[dsd_id]

# 4) Ver dimensões e codelists
dims = [c.id for c in dsd.dimensions.components]
print("Dimensões:", dims)

for comp in dsd.dimensions.components:
    cl = getattr(comp.local_representation, "enumerated", None)
    if cl is not None:
        # Recuperar a codelist completa do StructureMessage
        cl_obj = sm_dsd.codelists[cl.id]
        print(f"{comp.id} → codelist: {cl.id} ({len(cl_obj.items)} códigos)")
        # amostra de códigos
        for code_id, code in list(cl_obj.items)[:20]:
            print(" ", code_id, "-", code.name)
```
