## Etapa final: requisição do dataset

```python
import sdmx
import pandas as pd



# Cliente IMF
client = sdmx.Client("IMF_DATA")


key = 'BRA+USA.FL_S13_POGDP_PT+FL_PS_POGDP_PT.A'

# Parâmetros de tempo e atributos
params = dict(
    startPeriod="2000"
)

# Baixar os dados
data_msg = client.data("GDD", key=key, params=params, dsd=estrutura)

# Converter para pandas DataFrame
df = sdmx.to_pandas(
    data_msg, 
    #datetime="TIME_PERIOD"
    )

df = df.reset_index()

display(df.head())
```
